```python
import streamlit as st
import requests
import pandas as pd
import plotly.graph_objs as go

API_KEY = st.secrets["api"]["iex_key"]
API_BASE_URL = "https://cloud.iexapis.com/stable/"

def get_stock_data(symbol, time_range="1y"):
    params = {"token": API_KEY}
    response = requests.get(f"{API_BASE_URL}stock/{symbol}/chart/{time_range}", params=params)
    data = response.json()

    if "error" in data:
        st.error(f"Error: {data['error']}")
        return None

    stock_data = pd.DataFrame(data)
    stock_data["date"] = pd.to_datetime(stock_data["date"])
    stock_data.set_index("date", inplace=True)
    # stock_data = stock_data[["open", "high", "low", "close", "volume"]]
    # stock_data.columns = ["Open", "High", "Low", "Close", "Volume"]
    return stock_data

df = get_stock_data("AAPL", time_range="1y")
df.had()
```
